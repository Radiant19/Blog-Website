Reshape, Stacking, squeezing and unsqueezing tensors:
- Reshaping - reshapes an input tensors to a defined shape
- View - Return a view of an input tensor of certain shape but keep the same memory as original tensor
- Stacking - combine multiple tensors on top of each other (vstack) or side by side (hstack)
- Squeeze - removes all 1 dimension from tensor
- unsqueeze - add 1 dimension to a target tensor
- permute- return a view of the input with dimension permuted(swapped) in a certain way

```python
x = torch.arange(1.,10.)
x, x.shape
```

```python
x_reshaped = x.reshape(1,9)
x_reshaped, x_reshaped.shape
```

Reproducibility (trying to take random out of random)

In short how nn learns:

`start with random numbers -> tensor operations -> update random no to try and make them better representation of data again agian and again`

To reduce the randomnees in nural network and pytorch comes the concepy of random seed

essential what the seed does is'flavour' the randomness

```python
random_tensor_A = torch.rand(3,4)
random_tensor_B = torch.rand(3,4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
#Using Seed produce reproducible tensor

Random_Seed = 42
torch.manual_seed(Random_Seed)
random_tensor_C = torch.rand(3,4)

torch.manual_seed(Random_Seed)
random_tensor_D = torch.rand(3,4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
